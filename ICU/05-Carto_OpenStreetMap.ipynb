# Calcul

```python
# Fonction pour définir le shape (cercle) d'étude

def definir_cercle_en_polygone_geo(latitude, longitude, radius, nb_points=100):
    # Définir les systèmes de coordonnées (CRS)
    crs_wgs84 = CRS("EPSG:4326")  # WGS 84
    crs_utm = CRS("EPSG:32631")  # UTM zone 31N (Toulouse)

    # Créer un transformer pour la conversion entre WGS84 et UTM
    transformer_to_utm = Transformer.from_crs(crs_wgs84, crs_utm, always_xy=True)
    transformer_to_wgs84 = Transformer.from_crs(crs_utm, crs_wgs84, always_xy=True)

    # Convertir le centre du cercle en coordonnées UTM
    x_centre, y_centre = transformer_to_utm.transform(longitude, latitude)

    # Créer un cercle en UTM
    cercle_utm = Point(x_centre, y_centre).buffer(radius, resolution=nb_points)  # Le rayon est en mètres

    # Reprojeter le cercle en coordonnées géographiques
    cercle_geo_coords = [
        transformer_to_wgs84.transform(x, y)
        for x, y in np.array(cercle_utm.exterior.coords)
    ]

    # Créer un polygone en utilisant les coordonnées reprojetées
    cercle_geo = Polygon(cercle_geo_coords)

    # Retourner le polygone
    return cercle_geo

# Rayon en mètres
radius = 300  # 300m

# Calcul de la surface
surface_totale = math.pi * (radius ** 2)

# Définir le cercle comme un polygone géographique
cercle_polygone_geo = definir_cercle_en_polygone_geo(Latitude, Longitude, radius)

# Convertir le polygone en GeoJSON
cercle_geojson = json.dumps(mapping(cercle_polygone_geo), indent=2)

geometry_carmes = cercle_geojson

# Convertir la chaîne JSON en un dictionnaire Python
geometry_dict = json.loads(geometry_carmes)

# Convertir le dictionnaire en un objet Shapely Geometry
polygon = shape(geometry_dict)

# Créer un GeoDataFrame avec la géométrie
custom_shape = gpd.GeoDataFrame({'geometry': [polygon]}, crs='EPSG:4326')

#Batiments
try:
    # Récupérer les géométries selon les tags spécifiés
    geometries_bat = ox.features_from_polygon(custom_shape.geometry.iloc[0], tags=tag_b)

    if geometries_bat.empty:
        print("Aucun batiment trouvé.")
        # Créer un GeoDataFrame vide avec la même structure que les résultats attendus
        geometries_bat = gpd.GeoDataFrame(columns=['geometry'], geometry='geometry', crs="EPSG:4326")

except Exception as e:
    print(f"Erreur: {e}")
    # Créer un GeoDataFrame vide avec la même structure que les résultats attendus en cas d'erreur
    geometries_bat = gpd.GeoDataFrame(columns=['geometry'], geometry='geometry', crs="EPSG:4326")

# Eau
try:
    # Récupérer les géométries selon les tags spécifiés
    geometries_eau = ox.features_from_polygon(custom_shape.geometry.iloc[0], tags=tag_e)

    if geometries_eau.empty:
        print("Aucun élément eau trouvé.")
        # Créer un GeoDataFrame vide avec la même structure que les résultats attendus
        geometries_eau = gpd.GeoDataFrame(columns=['geometry'], geometry='geometry', crs="EPSG:4326")

except Exception as e:
    print(f"Erreur: {e}")
    # Créer un GeoDataFrame vide avec la même structure que les résultats attendus en cas d'erreur
    geometries_eau = gpd.GeoDataFrame(columns=['geometry'], geometry='geometry', crs="EPSG:4326")

# Espaces Verts
try:
    # Récupérer les géométries selon les tags spécifiés
    geometries_vert = ox.features_from_polygon(custom_shape.geometry.iloc[0], tags=tag_v)

    if geometries_vert.empty:
        print("Aucun espaces verts trouvé.")
       # Créer un GeoDataFrame vide avec la même structure que les résultats attendus
        geometries_vert = gpd.GeoDataFrame(columns=['geometry'], geometry='geometry', crs="EPSG:4326")

except Exception as e:
    print(f"Erreur: {e}")
    # Créer un GeoDataFrame vide avec la même structure que les résultats attendus en cas d'erreur
    geometries_vert = gpd.GeoDataFrame(columns=['geometry'], geometry='geometry', crs="EPSG:4326")

# Arbres unique
try:
    # Récupérer les géométries selon les tags spécifiés
    geometries_arb = ox.features_from_polygon(custom_shape.geometry.iloc[0], tags=tag_a)

    if geometries_arb.empty:
        print("Aucun arbre trouvé.")
        # Créer un GeoDataFrame vide avec la même structure que les résultats attendus
        geometries_arb = gpd.GeoDataFrame(columns=['geometry'], geometry='geometry', crs="EPSG:4326")

except Exception as e:
    print(f"Erreur: {e}")
    # Créer un GeoDataFrame vide avec la même structure que les résultats attendus en cas d'erreur
    geometries_arb = gpd.GeoDataFrame(columns=['geometry'], geometry='geometry', crs="EPSG:4326")

# Pieton
try:
    # Récupérer les géométries selon les tags spécifiés
    geometries_pieton = ox.features_from_polygon(custom_shape.geometry.iloc[0], tags=tag_p)

    if geometries_pieton.empty:
        print("Aucune zone piètonne trouvée.")
        # Créer un GeoDataFrame vide avec la même structure que les résultats attendus
        geometries_pieton = gpd.GeoDataFrame(columns=['geometry'], geometry='geometry', crs="EPSG:4326")

except Exception as e:
    print(f"Erreur: {e}")
    # Créer un GeoDataFrame vide avec la même structure que les résultats attendus en cas d'erreur
    geometries_pieton = gpd.GeoDataFrame(columns=['geometry'], geometry='geometry', crs="EPSG:4326")

# Affichage du nombre de bâtiments trouvés
print(f"Nombre d'éléments B: {len(geometries_bat)}")
print(f"Nombre d'éléments E: {len(geometries_eau)}")
print(f"Nombre d'éléments V: {len(geometries_vert)}")
print(f"Nombre d'éléments A: {len(geometries_arb)}")
print(f"Nombre d'éléments P: {len(geometries_pieton)}")

# Conversion des résultats en GeoDataFrame pour faciliter les manipulations
gdf_bat = gpd.GeoDataFrame(geometries_bat, geometry='geometry')
gdf_eau = gpd.GeoDataFrame(geometries_eau, geometry='geometry')
gdf_vert = gpd.GeoDataFrame(geometries_vert, geometry='geometry')
gdf_pieton = gpd.GeoDataFrame(geometries_pieton, geometry='geometry')

# Créer une fonction pour découper les éléments à la limite du polygone
def clip_element(element, polygon):
    return element.intersection(polygon)

# Appliquer la fonction à chaque élément en utilisant .loc pour éviter l'avertissement
gdf_bat.loc[:, 'geometry'] = gdf_bat['geometry'].apply(lambda x: clip_element(x, custom_shape.geometry.iloc[0]))
gdf_eau.loc[:, 'geometry'] = gdf_eau['geometry'].apply(lambda x: clip_element(x, custom_shape.geometry.iloc[0]))
gdf_vert.loc[:, 'geometry'] = gdf_vert['geometry'].apply(lambda x: clip_element(x, custom_shape.geometry.iloc[0]))
gdf_pieton.loc[:, 'geometry'] = gdf_pieton['geometry'].apply(lambda x: clip_element(x, custom_shape.geometry.iloc[0]))

# Supprimer les éléments qui n'ont pas d'intersection avec le polygone
gdf_bat = gdf_bat[~gdf_bat.is_empty]
gdf_eau = gdf_eau[~gdf_eau.is_empty]
gdf_vert = gdf_vert[~gdf_vert.is_empty]
gdf_pieton = gdf_pieton[~gdf_pieton.is_empty]

# Reprojeter le GeoDataFrame vers un CRS projeté (EPSG:2154) tout en travaillant sur une copie
g_bat = gdf_bat.to_crs(epsg=2154).copy()
g_eau = gdf_eau.to_crs(epsg=2154).copy()
g_vert = gdf_vert.to_crs(epsg=2154).copy()
g_pieton = gdf_pieton.to_crs(epsg=2154).copy()

# Calculer les surfaces
g_bat['surface_m2'] = g_bat.geometry.area
Surface_Total_B = g_bat['surface_m2'].sum()
g_eau['surface_m2'] = g_eau.geometry.area
Surface_Total_E = g_eau['surface_m2'].sum()
g_vert['surface_m2'] = g_vert.geometry.area
Surface_Total_V = g_vert['surface_m2'].sum()
g_pieton['surface_m2'] = g_pieton.geometry.area
Surface_Total_P = g_pieton['surface_m2'].sum()

# Hauteur building
# Vérifier si la colonne 'height' existe
if 'height' not in geometries_bat.columns:
    geometries_bat['height'] = np.nan  # Créer la colonne 'height' remplie de NaN
# Vérifier si la colonne 'building:levels' existe
if 'building:levels' not in geometries_bat.columns:
    geometries_bat['building:levels'] = np.nan  # Créer la colonne 'height' remplie de NaN

# Filtrer pour ne garder que les colonnes d'intérêt, en créant une copie explicite
geometries_bat_h = geometries_bat[['geometry', 'height', 'building:levels']].copy()

# Calculer la hauteur des bâtiments en utilisant les tags 'height' ou 'building:levels'
# Si 'height' n'est pas disponible, on peut estimer la hauteur via 'building:levels' (chaque niveau ≈ 3 mètres)
def estimate_height(row):
    try:
        if pd.notnull(row['height']):
            return float(row['height'])
        elif pd.notnull(row['building:levels']):
            return float(row['building:levels']) * 3  # Estimation : 3 mètres par étage
        else:
            return 3
    except ValueError:
        return 3


geometries_bat_h['calculated_height'] = geometries_bat_h.apply(estimate_height, axis=1)

# Calculer moyenne hauteur des batiments
Hauteur = round(geometries_bat_h['calculated_height'].mean(),2)

# Calculer surface sans type
surface_totale_Na = surface_totale -(Surface_Total_B + Surface_Total_E + Surface_Total_V + Surface_Total_P)

print(f'Surface Bâtiments: {round(Surface_Total_B,2)} m²')
print(f'Hauteur moyenne des batiments du quartier: {Hauteur} m')
print(f'Surface Zones Humides: {round(Surface_Total_E,2)} m²')
print(f'Surface Espaces Verts: {round(Surface_Total_V,2)} m²')
print(f'Surface Zones Piétonnes + Parking: {round(Surface_Total_P,2)} m²')
print(f'Surface Diverses: {round(surface_totale_Na ,2)} m²')
```

```python
# Extraire le centre du polygone pour centrer la carte Folium
map_folium = folium.Map(location=[Latitude, Longitude], zoom_start=16)

# Ajouter les polygones des bâtiments à la carte Folium
def add_geometries_to_map(gdf,g_utm, map_folium, color, fill_color):
    for _, row in gdf.iterrows():
        # Extraire la surface du bâtiment
        element_type, osmid = row.name
        surface = round(g_utm.loc[(element_type, osmid), 'surface_m2'],2) #if 'surface_m2' in row else None

        # Convertir la géométrie en GeoJSON
        geo_json = folium.GeoJson(
            row['geometry'],
            name=f'Surface: {surface}',
            style_function=lambda feature, color=color, fill_color=fill_color: {
                'color': color,
                'weight': 0.2,
                'fillColor': fill_color,
                'fillOpacity': 0.4
            }
        )

        # Ajouter un popup avec la surface
        popup = folium.Popup(f'Surface: {surface} m²',
                             max_width=150,  # Ajustez la largeur maximale du popup
                             parse_html=True) #if surface else None
        #if popup:
        geo_json.add_child(popup)

        # Ajouter le polygone à la carte
        geo_json.add_to(map_folium)


add_geometries_to_map(gdf_vert, g_vert, map_folium, 'lightgreen', 'green')
add_geometries_to_map(gdf_bat, g_bat, map_folium, 'grey', 'orange')
add_geometries_to_map(gdf_eau, g_eau, map_folium, 'blue', 'lightblue')
add_geometries_to_map(gdf_pieton, g_pieton, map_folium, 'lightgrey', 'black' )

popup_content = f"""
Station name: {station_name}<br>
Surface Bâtiments: {round(Surface_Total_B, 2)} m²<br>
Hauteur moyenne des bâtiments: {Hauteur} m<br>
Surface Zones Humides: {round(Surface_Total_E, 2)} m²<br>
Surface Espaces Verts: {round(Surface_Total_V, 2)} m²<br>
Nombre d'arbres: {len(geometries_arb)}<br>
Surface Zones Piétonnes + Parking: {round(Surface_Total_P, 2)} m²<br>
Surface diverses (non categorisée): {round(surface_totale_Na, 2)} m²<br>
Population_totale_300m:  {densité}<br>
Nb entreprises: {Ent_count}
"""

# Ajouter le marqueur au centre du cercle
folium.Marker(
    location=[Latitude, Longitude],
    popup=folium.Popup(
        popup_content,
        max_width=250  # Ajustez la largeur maximale du popup
    ),
    icon=folium.Icon(icon='temperature-arrow-up', prefix='fa')  # Utilisation de l'icône de thermomètre depuis Font Awesome
).add_to(map_folium)

# Ajouter le cercle de 300 mètres de rayon
folium.Circle(
    location=[Latitude, Longitude],
    radius=300,  # Rayon en mètres
    color='green',
    fill=False,

).add_to(map_folium)

# Sauvegarder la carte Folium
file_path_map = os.path.join(path_carto, f'carte_{id}.html')
map_folium.save(file_path_map)

print(f"Carte Folium enregistrée sous {file_path_map}.")
map_folium
```
